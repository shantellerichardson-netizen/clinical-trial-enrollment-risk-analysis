```python
import pandas as pd

df = pd.read_csv("/content/AERO-BirdsEye-Data.csv")

print(df.head())
print(df.info())
print(df.describe())
```

```python
#problem 1
df = df.dropna(subset=["Phase"])
df["Phase"] = df["Phase"].str.strip()

df.groupby("Phase")["Enrollment"].mean().sort_values()
```

The later the phases the higher enrollment across clinical trials.
Later phases of trials show significantly higher enrollment, which could indicate either increased operation scale or resource requirements in advance stages of development. Phase 3 shows the highest enrollment rate out of all.


"Enrollment size increases substantially across clinical trial phases, with Phase 3 trials averaging nearly 800 participants compared to approximately 50 in Phase 1. This reflects the growing scale and resource requirements of later-stage trials, where broader populations are needed to validate efficacy and safety."

```python
#problem 2
status_counts = df["Status"].value_counts(normalize=True) * 100
print(status_counts)
```

```python
pd.crosstab(df['Phase'], df['Status'], normalize='index') * 100
#print("Completed")
#print("Terminated")
```

76.75% of clinical trials are completed while 9.38% are terminated showing that the majority of clinical trials across the board are adequately following requirements and regulations. (mangerial insight)

“Approximately 77% of clinical trials reach completion, while nearly 10% are terminated early. This highlights a meaningful level of operational and financial risk in the R&D pipeline, emphasizing the importance of effective trial design and resource allocation.”

```python
status_by_phase = pd.crosstab(df['Phase'], df['Status'], normalize='index') * 100
print(status_by_phase[['Completed', 'Terminated']])
```

“Completion rates vary across clinical trial phases, with Phase 4 and Phase 3 trials demonstrating the highest likelihood of completion. In contrast, transitional phases such as Phase 1/Phase 2 and Phase 2/Phase 3 exhibit notably lower completion rates and higher termination rates. This suggests that these intermediate stages represent critical decision points in the clinical development process, where trials face increased scientific and operational uncertainty. From a managerial perspective, this highlights the importance of careful evaluation and resource allocation during transitional phases, as they carry a higher risk of failure. These findings indicate that while later-stage trials require greater investment due to larger enrollment sizes, the greatest risk of failure occurs earlier in the development pipeline.”
